# 04-2 matplotlib 라이브러리 자유자재로 사용하기

## 기초 그래프 그리기

### 기초 그래프 그리기 - 히스토그램, 산점도 그래프, 박스 그래프

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
type(tips)
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

```python
axes1.hist(tips['total_bill'], bins = 10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
tips.head()
```

```python
tips_female = tips[tips['sex'] == 'Female']['tip']
```

```python
tips_female
```

```python
boxplot = plt.figure()
```

```python
axes1 = boxplot.add_subplot(1,1,1,)
```

```python
axes1.boxplot([
               tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']
               ] )
```

```python
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
boxplot
```

```python
tips_male = tips[tips['sex'] == 'male']['tip']
```

# 04-3  seaboron 라이브러리 자유자재로 사용하기

### 단변량 그래프 그리기 - 히스토그램

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = sns.displot(tips['total_bill'])
ax.set_titles('Total Bill Histogram with Density Plot')
```

```python
ax = sns.displot(tips['total_bill'], kde=False)
ax.set_xlabel('Total Bill')
```
