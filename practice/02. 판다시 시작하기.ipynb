## 데이터 추출하기 - 슬라이싱 구문, range 메서드

### (1)  슬라이싱 구문으로 데이터 추출하기

```python
subset = df.loc[:, ['year', 'pop'] ] 
print(subset.head() )
```

```python
subset = df.iloc[ : , [2, 4, -1] ]
print(subset.head())
```

### (2) range 메서드로 데이터 추출하기

```python
small_range = list(range(5))
print(small_range)
```

```python
subset = df.iloc[ : , small_range] 
```

```python
print(subset.head())
```
