## 데이터프레임 만들기

```python
scientists = pd.DataFrame({
    'Name' : ['Rosaline Franklin', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]}
)
```

```python
print(scientists)
```

```python
scientists_addindex = pd.DataFrame({
    'Name' : ['Rosaline Franklin', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]}, index = ['Rosaline Frankline', 'William Gosset'])
```

```python
print(scientists_addindex)
```

```python
scientists_addindex2 = pd.DataFrame(data = {
    'Name' : ['Rosaline Franklin', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]}, index = ['Rosaline Frankline', 'William Gosset'])
```

```python
print(scientists_addindex2)
```

```python
scientists_addindex_columns = pd.DataFrame(data = {
    'Name' : ['Rosaline Franklin', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]}, 
    index = ['Rosaline Frankline', 'William Gosset'],
    columns = ['Occupation', 'Born', 'Age', 'Died'])
```
